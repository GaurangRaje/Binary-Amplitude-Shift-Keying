# Importing the libraries

```python
import numpy as np
from scipy import signal
import pandas as pd
import matplotlib.pyplot as plt
```

# Bipolar Amplitude Sift Keying (BASK)

## Carrier and Modulating wave parameters

```python
fc = int(input("Enter frequency of carrier signal: "))
fm = int(input("Enter frequency of modulating signal: "))
A = int(input("Enter amplitude of carrier signal: "))
```

## Sampling rate

```python
t = np.linspace(start=0, stop=1, num=10000, endpoint=True)
```

## Carrier wave
## Vcarrier = Asin(2Ï€ft)

```python
carrier = np.array(A*np.sin(2*np.pi*fc*t)).T
```

## Modulating wave

```python
b = np.array(signal.square(2*np.pi*fm*t)).T
u = [int(x+1) if x == -1 else int(x) for x in b]
```

## Modulated signal BASK
## Vbask = u(t)*Vcarrier

```python
BASK = np.array(u*carrier).T
```

# Plotting the results

```python
fig, axes = plt.subplots(nrows=3, ncols=1)
axes[0].plot(t, carrier)
axes[1].plot(t, u)
axes[2].plot(t, BASK)

axes[0].set_title('Carrier wave')
axes[1].set_title('Modulating signal (Unipolar)')
axes[2].set_title('BASK signal')

plt.tight_layout()

fig.set_figheight(12)
fig.set_figwidth(20)

axes[0].set_xlabel('time')
axes[0].set_ylabel('Amplitude')
axes[1].set_xlabel('time')
axes[1].set_ylabel('Amplitude')
axes[2].set_xlabel('time')
axes[2].set_ylabel('Amplitude')

axes[0].grid(which='major', axis='both')
axes[1].grid(which='major', axis='both')
axes[2].grid(which='major', axis='both')
```
